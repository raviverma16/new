```python
import tabula
```

```python
pdf_path = "https://www.lseg.com/curve-files/2021/09/TradingSummary_2021-09-10.pdf"

dfs = tabula.read_pdf(pdf_path, pages="all")
```

```python
eui = dfs[6]
stl = dfs[7]
son = dfs[9]
stg = dfs[15]
sna = dfs[17]
osn = dfs[11]
osa = dfs[13]
```

**EURIBOR**

```python
eui['Three Month Euribor Future']
eui_1 = eui['Three Month Euribor Future']

eui_1 = eui_1.drop(index=[0,1,2,3,4])

eui_1
```

```python
eui_code = eui_1.str[:5]
eui_date = eui_1.str[6:17]
eui_dsp = eui_1.str[17:]
```

```python
import pandas as pd
```

```python
euribor = pd.DataFrame()
euribor["Instrument"] = eui_code
euribor["Expiration Date"] = eui_date
euribor["DSP"] = eui_dsp
euribor = euribor.reset_index()
euribor = euribor.drop(columns=['index'])
euribor = euribor.drop(index=[28])
euribor.tail()
```

**SONIA**

```python
son.head()
```

```python
son_1 = son.drop(index=[0], columns=['Unnamed: 12', 'Unnamed: 11'])
son_1 = son_1.rename(columns=son_1.iloc[0])
son_1 = son_1.drop(index=[1])
son_1 = son_1.reset_index()
son_1 = son_1.drop(columns=['index'])

son_1
```

```python
sonia = pd.DataFrame()
sonia["Instrument"] = son_1["Instrument"]
sonia["Expiration Date"] = son_1["Expiration\rDate"]
sonia["DSP"] = son_1["DSP"]
sonia = sonia.drop(index=[29])
sonia.tail()
```

**SONIA_DNC**

```python
sna.head()
```

```python
sna_1 = sna.drop(index=[0], columns=['Unnamed: 12', 'Unnamed: 11'])
sna_1 = sna_1.rename(columns=sna_1.iloc[0])
sna_1 = sna_1.drop(index=[1])
sna_1 = sna_1.reset_index()
sna_1 = sna_1.drop(columns=['index'])

sna_1
```

```python
sonia_dnc = pd.DataFrame()
sonia_dnc["Instrument"] = sna_1["Instrument"]
sonia_dnc["Expiration Date"] = sna_1["Expiration\rDate"]
sonia_dnc["DSP"] = sna_1["DSP"]
sonia_dnc = sonia_dnc.drop(index=[29])
sonia_dnc.tail()
```

**STERLING**

```python
stl.head()
```

```python
stl_1 = stl.drop(index=[0], columns=['Unnamed: 12', 'Unnamed: 11'])
stl_1 = stl_1.rename(columns=stl_1.iloc[0])
stl_1 = stl_1.drop(index=[1])
stl_1 = stl_1.reset_index()
stl_1 = stl_1.drop(columns=['index'])

stl_1
```

```python
sterling = pd.DataFrame()
sterling["Instrument"] = stl_1["Instrument"]
sterling["Expiration Date"] = stl_1["Expiration\rDate"]
sterling["DSP"] = stl_1["DSP"]
sterling = sterling.drop(index=[26])
sterling.head()
```

**STERLING_DNC**

```python
stg.head()
```

```python
stg_1 = stg.drop(index=[0], columns=['Unnamed: 12', 'Unnamed: 11'])
stg_1 = stg_1.rename(columns=stg_1.iloc[0])
stg_1 = stg_1.drop(index=[1])
stg_1 = stg_1.reset_index()
stg_1 = stg_1.drop(columns=['index'])

stg_1
```

```python
sterling_dnc = pd.DataFrame()
sterling_dnc["Instrument"] = stg_1["Instrument"]
sterling_dnc["Expiration Date"] = stg_1["Expiration\rDate"]
sterling_dnc["DSP"] = stg_1["DSP"]
sterling_dnc = sterling_dnc.drop(index=[26])
sterling_dnc.tail()
```

```python
merged = pd.DataFrame()
merged = pd.concat([euribor, sonia, sonia_dnc, sterling, sterling_dnc])
merged = merged.reset_index()
merged = merged.drop(columns=['index'])
merged
```
